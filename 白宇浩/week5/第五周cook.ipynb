```python
import fasttext

model = fasttext.train_unsupervised('cooking.stackexchange.txt')
```

```python
print(model.words)
```

```python
print(model.labels)
```

```python
model.save_model("cook.bin")
```

```python
model = fasttext.load_model("cook.bin")
```

```python
word_vector = model.get_word_vector("and") 
print(f"单词 'and' 的词向量维度：{word_vector.shape}")

# 2. 查找最近邻词
neighbors = model.get_nearest_neighbors("and", k=5)  
print("与 'and' 最接近的词及相似度：")
for score, word in neighbors:
    print(f"{word}: {score:.3f}")
```
