```python
#数据分词
import jieba

with open('gzr.txt','r',encoding='utf-8') as f:
    lines = f.read()

with open('gzr_fc.txt','w',encoding='utf-8') as f:
    f.write(' '.join(jieba.cut(lines)))
```

```python
import numpy as np
import torch
import torch.nn as nn
from torch.utils.tensorboard import SummaryWriter

# 加载FastText模型
wv_file = 'gzr_fc.bin'
word_vectors = fasttext.load_model(wv_file)

# 获取词汇表
vocab = word_vectors.get_words()  # 直接获取所有单词列表

# 构建嵌入矩阵
embedding_dim = word_vectors.get_dimension()
embedding_matrix = np.zeros((len(vocab), embedding_dim))
for i, word in enumerate(vocab):
    embedding_matrix[i] = word_vectors.get_word_vector(word)  # 使用单词字符串作为输入

# 转换为PyTorch Embedding
embedding = nn.Embedding.from_pretrained(torch.FloatTensor(embedding_matrix))

# 写入TensorBoard
writer = SummaryWriter()
meta = vocab[:100]  # 直接取前100个单词作为元数据
writer.add_embedding(embedding.weight[:100],metadata=meta)
writer.close()
```
