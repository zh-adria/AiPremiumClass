```python
from time import time

import torch
from torch import nn
from torch.utils.data import DataLoader
from torchvision import datasets
from torchvision.transforms import ToTensor,Lambda, Compose
import matplotlib.pyplot as plt

def creat_modle_datas():
    """
    从指定的库里面下载对应的模型训练数据
    :return:
    """
    train_data = datasets.FashionMNIST(
        root = "data",
        train = True,
        download=True,
        transform=ToTensor(),
    )
    test_data = datasets.FashionMNIST(
        root = "data",
        train=False,
        download=True,
        transform=ToTensor(),
    )
    return train_data, test_data

def creat_modle_KMNIST_datas():
    """
    从指定的库里面下载对应的模型训练数据
    :return:
    """
    train_data = datasets.KMNIST(
        root = "data",
        train = True,
        download=True,
        transform=ToTensor(),
    )
    test_data = datasets.KMNIST(
        root = "data",
        train=False,
        download=True,
        transform=ToTensor(),
    )
    return train_data, test_data


def load_model_data(train_date, test_data, bitch_size = 64):
    """
    对模型的数据进行加载处理
    :return:
    """
    # 创建数据加载器
    train_dataloader = DataLoader(train_date, batch_size= bitch_size, shuffle= True)
    test_data = DataLoader(test_data, batch_size= bitch_size, shuffle=True)
    return train_dataloader, test_data
class NeuralNetwork(nn.Module):
    def __int__(self):
        super(NeuralNetwork, self).__init__()

def sample_network():
    """
    :return:
    """
    x = torch.randn(10,784)
    # 线性层
    linear = nn.Linear(in_features=784, out_features=64, bias=True)
    act = nn.Sigmoid()
    # 隐藏层
    out = linear(x)
    out2 = act(out)
    # 输出层
    linear2 = nn.Linear(in_features=64, out_features=10, bias=True)
    out3 = linear2(out2)
    print(out3.shape)
    softmax = nn.Softmax(dim = 1)
    final = softmax(out3)
    print(final)

def chuanlian_model():
    """
    串联模型
    :return:
    """
    # 超参数
    lr = 1e-3
    epoches = 50
    #数据准备
    # train_data, testdata = creat_modle_datas()
    train_data, testdata = creat_modle_KMNIST_datas()
    training_data, testing_data = load_model_data(train_data, testdata)
    # 所有结构串联
    model =  nn.Sequential(
        nn.Linear(784, 64),
        nn.Sigmoid(),
        nn.Linear(64, 10)
    )
    # 损失函数
    loss_func = nn.CrossEntropyLoss()
    # 优化器
    opyimizer = torch.optim.SGD(model.parameters(), lr=lr)
    #
    # for epoch in range(epoches):
    #     for data,target in train_data:
    #         # 前向运算
    #         output = model(data.reshape(-1))
    #         # 计算损失
    #         loss = loss_func(output, torch.tensor(target))
    #         opyimizer.zero_grad()
    #         loss.backward()
    #         opyimizer.step()
    #        print("loss:", loss.item())
    for epoch in range(epoches):
        for data,target in training_data:
            output = model(data.reshape(-1,784))
            # 计算损失
            loss = loss_func(output, torch.tensor(target))
            opyimizer.zero_grad()
            loss.backward()
            opyimizer.step()
        print("loss:", loss.item())
    correct = 0
    total = 0
    with torch.no_grad():
        for data, target in testing_data:
            output = model(data.reshape(-1, 784))
            _,predicted = torch.max(output, 1)
            total += target.numel()
            correct += (predicted == target).sum().item()
    print(f"Accuracy:{correct/total *100}%")


def main():
    train_data, test_data = creat_modle_datas()
    load_model_data(train_data, test_data)
if __name__ == '__main__':
    # main()
    chuanlian_model()
```
