```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plot_x = np.linspace(-1,6,150)
plot_y = (plot_x -2.5) ** 2- 1
plt.plot(plot_x,plot_y)
plt.show()
```

```python
# 计算导数
def derivate(theta):
    return 2 * (theta - 2.5)

#  计算损失函数，也就是 函数预测值和实际值的偏差
def loss(theta):
    return (theta-2.5) **2 -1
```

```python
# 计算梯度
def gradient_descent(init_theta,eta,epsilon,num):
    # 收集theta 
    theta_his = [init_theta]
    # 初始化
    theta = init_theta
    
    iter_i = 0
    while iter_i < num:
        theta_his.append(theta)
        gradient = derivate(theta) # 参数值对应的梯度
        last_theta = theta
        theta = theta - eta * gradient #更新参数
        
        # 损失函数值
        lossv = loss(theta)

        # 中断循环的条件
        # if abs(gradient) < epsilon:
        #     break

        # 中断循环条件2
        # if abs(lossv) == 0:
        #     break

        # 中断循环条件 连续损失函数差小于一个阈值
        if abs(lossv - loss (last_theta)) < epsilon:
            break
        iter_i += 1
    return theta_his
```

```python
def plot_theta_history(theta_his):
    plt.plot(plot_x,plot_y)# 注释后可以只展示 梯度数据
    plt.plot(theta_his,loss(np.array(theta_his)),color = 'darkgreen',marker = '+')
    plt.xlabel('xlabel')
    plt.ylabel('ylabel')
    plt.title('1')
    plt.show()
```

```python
theta = 0.0 # 初始值
eta = 1.1 # 学习率，学习率越大，会出现类似震荡的横跳，有助于加速梯度收敛的速度，类似下山过程中大步迈步；这个值越大，后面的梯度会越少。
epsilon = 1e-8 # 阈值
num = 1e2 # 最大值
theta_his = gradient_descent(theta,eta,epsilon,num)

print(len(theta_his)-1) # 训练次数
# 绘制损失函数的变化趋势图
plot_theta_history(theta_his)
```
