```python
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.datasets import make_classification
import numpy as np
```

```python
X,y = make_classification(n_samples=1000,n_features=10)
# 快速生了100个样本，每一行都是一个样本，样本集合在X里面，有20个特征值，特质取值有正有负，接近正态分布
# print(X)
# print(y)
# print(X.shape)
# print(y.shape)

# 局部样本训练模型（过拟合模型）
# 新样本数据模型训练效果不好，泛化能力差；是否能对新的数据进行很好的训练 
# train_X,train_y,test_X,test_y = train_test_split(X,y)
train_x,test_x,train_y,test_y = train_test_split(X,y,test_size=0.2,shuffle=True) # 使用30%的数据进行训练
```

```python
# 假设X 是shape（n,n）
#  X.T shape(n,n)
# 假设theta 模型参数 shape(1,3)    [[w1,w2,w3]]

def forward(x,theta,bias):
    # 线性运算
    z = np.dot(theta,x.T)+bias
    # sigmid 函数
    y_hat= 1/ (1+np.exp(-z))
    return y_hat.squeeze()

def loss_fun(y, y_hat):
    e = 1e-8  # 防⽌y_hat计算值为0，添加的极⼩值epsilon
    return -np.mean(y * np.log(y_hat + e) + (1 - y) * np.log(1 - y_hat + e))

def calc_gradient(x,y,y_hat):
    m = x.shape[0]
    deltac_theta = np.dot((y_hat-y),x)/ m 
    delta_bias = np.mean(y_hat-y)
    return deltac_theta,delta_bias
```

```python
# 权重参数
theta= np.random.randn(1,X.shape[1])  #shape (1,10)
bias = 0
# 超参数
learnrate = 0.1 # 学习率
epochs = 5000 #训练次数

for i in range(epochs):
    y_hat = forward(train_X,theta,bias)
    # 计算损失
    loss0 = loss_fun(train_y,y_hat)
    
    dw,db = calc_gradient(train_X,train_y,y_hat)
    # 更新参数
    theta -= learnrate * dw
    bias -= learnrate * db

    if i%100 == 0:
        acc = np.mean(np.round(y_hat) == train_y)
        print('step:',i,'loss:',loss0,' acc:',acc)



```

```python
"""
1. 数据准备，参数初始化
2. 前向计算
3. 计算损失
4. 计算梯度
5. 更新参数
"""
import torch
from sklearn.datasets import load_iris
from sklearn.datasets import make_classification
# 超参数：学习率
learn_rate = 1e-3
# 1.1 数据准备
X, y = make_classification(n_features=10)
# 创建张量
tensor_x = torch.tensor(X, dtype=torch.float)
tensor_y = torch.tensor(y, dtype=torch.float)
# 1.2 创建参数并初始化
# x [100,10] * w[1,10]
w = torch.randn(1, 10, requires_grad=True) # 初始化参数w
b = torch.randn(1, requires_grad=True) # 初始化参数b
for i in range(5000):
    # 2. 前向运算
    r = torch.nn.functional.linear(tensor_x, w, b)
    r = torch.sigmoid(r)
    # 3. 计算损失
    loss = torch.nn.functional.binary_cross_entropy(r.squeeze(1), tensor_y, reduce=True)
    # 4.计算梯度
    loss.backward()
    # 5.参数更新
    with torch.autograd.no_grad(): # 关闭梯度计算跟踪
        w -= learn_rate * w.grad # 更新权重梯度
        w.grad.zero_() # 清空本次计算的梯度（因为梯度是累加计算，不清空就累
        b -= learn_rate * b.grad # 更新偏置项梯度
        b.grad.zero_() # 清空本次计算的梯度
    
    # item()张量转换python基本类型
    print(f'train loss:{loss.item():.4f}')
```
