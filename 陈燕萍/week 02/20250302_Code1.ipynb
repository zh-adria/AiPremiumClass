```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plot_x = np.linspace(-1, 6, 150)
print(plot_x)
```

```python
plot_y = (plot_x - 2.5) ** 2 - 1
```

```python
plt.plot(plot_x, plot_y)
plt.show()
```

```python
# 计算导数 theta是指模型中的参数
def derivate(theta):
    return 2*(theta-2.5)

# 计算损失函数
def loss(theta):
    try:
        return(theta-2.5)**2-1
    except:
        return float('inf') # 防止溢出 infinite极值；返回一个极大值
```

```python
theta = 0.0 # 参数初始值
eta = 0.01 # 学习率 可调整学习率，0.1，0.01
epsilon = 1e-8 # 精度
```

```python
gradient = derivate (theta) # 参数值对应的梯度
    
theta = theta - eta * gradient # 更新参数
    
theta # 运行后theta不断更新
```

```python
# 梯度下降
def gradient_descent(init_theta, eta, epsilon=1e-8):
    # 收集theta的值
    theta_history = [init_theta]
    theta = init_theta
    # 记录循环次数
    iter_i =0

    while iter_i < 1e4:
        gradient = derivate (theta) # 参数值对应的梯度
        # 记录上一次theta的值
        last_theta = theta
        # 更新参数值
        theta = theta - eta * gradient 
        # 记录theta
        theta_history.append(theta)
        # 损失函数值
        loss_v = loss (theta)
        # 中断循环条件 (连续两次损失函数差小于一个阈值)
        if abs(loss_v - loss(last_theta)) < epsilon: # 可以放大阈值 1e-7
            break

        iter_i +=1
    return theta_history

```

```python
print (theta)
print (loss(theta))
```

```python
def plot_theta_history(theta_history):
    plt.plot(plot_x, plot_y)
    plt.plot(theta_history, loss(np.array(theta_history)), color='r', marker='+')
    plt.show()
 
```

```python
theta = 0.0 # 参数初始值
eta = 0.005 # 学习率 可调整学习率，0.1，0.01，0.05；当学习率过大时0.8会出现震荡图形;通常不会太大
epsilon = 1e-8 # 精度
```

```python
theta_history = gradient_descent(theta, eta)
# 训练次数
len(theta_history) - 1
# 绘制损失函数变化图
plot_theta_history(theta_history)

print (len(theta_history) - 1)
print(theta_history[-1])
```
