```python
#初步实现
import torch
import torch.nn as nn  # 神经网络模块

#隐藏层

#线性层
linear1 = nn.Linear(784, 64, bias=True)  # 输入层，输出层，偏置
#激活层
# relu = nn.ReLU()  # 激活函数
act = nn.Sigmoid()  # 激活函数
#输出层
linear2 = nn.Linear(64, 10, bias=True)  # 输入层，输出层，偏置

#模拟输入
x = torch.randn(10, 784)
#前向传播
y = linear1(x)
print(y)
y = act(y)
print(y)
y = linear2(y)
print(y)
print(y.shape)

softmax = nn.Softmax(dim=1)  # 激活函数
final = softmax(y)  # 激活函数
print(final)
```

```python
#结构串联
model = nn.Sequential(
    nn.Linear(784, 64),
    nn.Sigmoid(),
    nn.Linear(64, 10)
)


loss_fn = nn.CrossEntropyLoss()  # 交叉熵损失函数
optimizer = torch.optim.SGD(model.parameters(), lr=0.01)  # SGD优化器

[param for param in model.parameters()]
```
