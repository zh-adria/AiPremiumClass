```python
# 梯度下降
import numpy as np
import matplotlib.pyplot as plt

plot_x = np.linspace(-1,6,150)
plot_y = (plot_x - 2.5) ** 2 - 1

theta = 0 # 初始值
eta = 1.1 # 学习率
epsilon = 1e-8 # 退出阈值


# 计算导数
def derivative(theta):
    return 2*(theta-2.5)
# 计算损失函数
def loss(theta):
    try:
        return (theta-2.5)**2 - 1
    except:
        return float('inf') # 返回一个极大值，防止溢出

def gradient_descent(init_theta, eta, epsilon=1e-8):
    # 收集theta的值
    theta_history = [init_theta]
    theta = init_theta
    # 记录循环次数
    iter_i = 0

    while iter_i < 10:
        gradient = derivative(theta) # 参数值对应的梯度
        # 记录上一次theta值
        last_theta = theta
        # 更新参数
        theta = theta - eta * gradient 
        # 记录theta
        theta_history.append(theta)
        # 损失函数值
        loss_v = loss(theta)
        # 循环终止条件
        if abs(loss_v - loss(last_theta)) < epsilon:
            break
        
        iter_i += 1
    return theta_history

def plot_theta_history(theta_history):
    plt.plot(plot_x,plot_y)
    plt.plot(theta_history,loss(np.array(theta_history)),color='r',marker='+')
    plt.show()


theta_history = gradient_descent(theta, eta)
print(len(theta_history))
print(theta_history[-1])
plot_theta_history(theta_history)
```
