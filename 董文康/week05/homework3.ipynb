```python
import jieba
```

```python
with open('./xyj.txt', 'r') as f:
    lines = f.read()

with open('./xyj_c.txt', 'w') as f:
    f.write(' '.join(jieba.cut(lines)))
```

```python
import fasttext
```

```python
model = fasttext.train_unsupervised('xyj_c.txt')
```

```python
print('文档词汇表：', model.get_words())
```
