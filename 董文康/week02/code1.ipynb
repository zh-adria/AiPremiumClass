```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plot_x = np.linspace(-1, 6, 141)
print(plot_x)
```

```python
plot_y = (plot_x - 2.5) ** 2 - 1
```

```python
plt.plot(plot_x, plot_y)
plt.show()
```

```python
# 计算导数
def derivate(theta):
    return 2 * (theta - 2.5)  # 对theta求导


# 损失函数
def loss(theta):
    return (theta - 2.5) ** 2 - 1


# 梯度下降
def gradient_descent(init_theta, eta, epsilon=1e-8):
    theta = init_theta
    theta_history = [theta]
    theta_history.append(init_theta)
    iteration = 0
    while iteration < 80:
        gradient = derivate(theta)
        last_theta = theta
        theta = theta - eta * gradient
        theta_history.append(theta)
        if abs(loss(theta) - loss(last_theta)) < epsilon:
            break
        iteration += 1
    return theta_history
```

```python
def plot_theta_history(theta_history):
    plt.plot(plot_x, loss(plot_x))
    plt.plot(np.array(theta_history), loss(np.array(theta_history)), color='r', marker='+')
    plt.show()
```

```python
theta = 0.0 # 初始值
eta = 0.08  # 学习率
theta_history = gradient_descent(theta, eta)
plot_theta_history(theta_history)
print(theta_history)
```
