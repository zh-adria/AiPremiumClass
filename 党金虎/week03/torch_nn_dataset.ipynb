```python
import torch 
from torchvision.datasets import  FashionMNIST
from torchvision.transforms import ToTensor
from torch.utils.data import DataLoader


# 加载数据集
train_data = FashionMNIST(root='./党金虎/week06/data', train=True, download=True,transform=ToTensor())  
# 测试数据集
test_data = FashionMNIST(root='./党金虎/week06/data', train=False, download=True,transform=ToTensor())

print('train_data:', len(train_data))
print('test_data:', len(test_data))
print(train_data[1])

print(train_data[1][0])


```

```python
import matplotlib.pyplot as plt
img,clzz = train_data[59999]
plt.imshow(img[0],cmap='gray')
plt.show()
print(clzz)
```

```python
# 打印图像和标签信息
print(f"图像形状: {img.shape}")  # 输出: torch.Size([1, 28, 28])
print(f"类别标签: {clzz}")       # 输出: 标签值（如 5）

# 可视化图像
plt.imshow(img.squeeze(), cmap='gray')  # 去掉批次维度并显示灰度图像
plt.title(f"Label: {clzz}")
plt.show()

# 提取所有标签
labels = [label for _, label in train_data]

# 打印所有可能的标签值
unique_labels = set(labels)
print(f"数据集中所有可能的标签值: {unique_labels}")
```
