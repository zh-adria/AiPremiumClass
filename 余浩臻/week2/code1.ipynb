```python
#梯度下降
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 生成一个包含 150 个元素的一维数组，元素均匀分布在 -1 到 6 这个区间内,递增数列
plot_x=np.linspace(-1,6,150)
# 打印生成的数组
print(plot_x)
```

```python
# 计算每个 plot_x 对应的 y 值，使用公式 (x - 2.5)^2 - 1
plot_y=(plot_x-2.5)**2-1
```

```python
plt.plot(plot_x, plot_y)
plt.show()
```

```python
#计算导数
def derivative(theta):
    return (theta-2.5)*2

#计算损失函数
def loss(theta):
    try:
        return (theta-2.5)**2-1
    except:
        return float('inf')#防止溢出，返回一个无穷大的数
```

```python
def gradient_descent(init_theta,eta,epsilon=1e-8): #梯度下降函数
    #收集theata的值
    theta_history=[init_theta]
    theta=init_theta
    #代码实现了一个基于梯度下降算法来寻找损失函数最小值的迭代过程。


    #记录循环次数
    iter_i=0
    #原本while True
    while iter_i<1e4:
        gradient=derivative(theta) #参数值对应的梯度

        #记录上一次的theta和损失值
        theta_last=theta
        
        #更新参数值
        theta=theta - eta * gradient 

        #记录theta
        theta_history.append(theta)

        loss_v=loss(theta) #更新损失值

        #连续损失函数的差值小于一个阈值，中断循环##
        if abs(loss_v-loss(theta_last))<epsilon:
            break

        iter_i=iter_i+1
    return theta_history

        #if abs(gradient)<1e-8: #如果损失值小于阈值，则停止迭代
            #break

        #损失函数很难等于0，只能不断逼近与0的值
        #if abs(loss_v)==0:
            #break
```

```python
#绘制梯度函数
def plot_theta_history(theta_history):
    plt.plot(plot_x, plot_y)
    plt.plot(theta_history,loss(np.array(theta_history)),color='red',marker='+')
    plt.show()
```

```python
theta=0.0 #初始值
eta=1.01 #学习率 0.1 45次学习 0.01 423次学习 
#1.1学习率会无限制运行，防止此情况，在原来的while true死循环处设置循环次数的阈值
#不会收敛，会不断发散，因为范围是1e4，所以执行到10001次后跳出循环
epsilon=1e-8 #精度
```

```python
theta_history=gradient_descent(theta,eta)

#训练次数
len(theta_history)-1

#绘制
plot_theta_history(theta_history)

print(len(theta_history))
```
