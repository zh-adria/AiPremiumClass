```python
# 读取文本
# 训练词向量 fasttext word2vec
# 计算词汇间的相关度
# tensorboard可视化

import fasttext
# import requests

# resp = requests.get('https://www.isyd.net/books/3/3.txt')
# resp.encoding = 'utf-8'
# lines = resp.iter_lines(chunk_size=1024)
# lines = [line.decode('utf-8') for line in lines if line]

model = fasttext.train_unsupervised('/kaggle/input/men-are-brothers-word-embedding/men-are-brothers.txt', model='skipgram')
```

```python
nns = model.get_nearest_neighbors('西边')
an = model.get_analogies('西边','南边','西山')
wv = model.get_word_vector('西边')
print(nns)
print(an)
print(wv)
```
